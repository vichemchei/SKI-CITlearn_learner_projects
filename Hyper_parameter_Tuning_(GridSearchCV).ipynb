```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(iris.data, iris.target, test_size= 0.3, random_state= 20)
```

```python
from sklearn.svm import SVC
model = SVC(kernel='rbf',C = 30, gamma= 'auto')
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(SVC(kernel='linear',C=20, gamma='auto'),iris.data,iris.target, cv=5)
```

```python
cross_val_score(SVC(kernel='rbf',C=20, gamma='auto'),iris.data,iris.target, cv=5)
```

```python
cross_val_score(SVC(kernel='rbf',C=10, gamma='auto'),iris.data,iris.target, cv=5)
```

```python
kernels = ['rbf','linear']
C = [1,10,20,30]

avg_score = {}
for kval in kernels:
    for cval in C:
        cv_score = cross_val_score(SVC(kernel=kval,C= cval, gamma='auto'),iris.data,iris.target, cv =5 )
        avg_score[kval +'_' + str(cval)] = np.average(cv_score)

avg_score
```

```python
from sklearn.model_selection import GridSearchCV

gsv = GridSearchCV(SVC(gamma='auto'),{
    'C': [1,10,20,30],
    'kernel': ['rbf', 'linear']
}, cv = 5, return_train_score= False)

gsv.fit(iris.data,iris.target)
```

### RandomizedSearchCV

### Choosing the right model

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
model_params = {
    'svm': {
        'model': svm.SVC(gamma='auto'),
        'params': {
            'C': [1, 10, 20],
            'kernel': ['rbf', 'linear']
        }
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params': {
            'n_estimators': [1, 5, 10]
        }
    },
    'logistic_regression': {
        'model': LogisticRegression(solver='liblinear', multi_class='auto'),
        'params': {
            'C': [1, 5, 10]
        }
    }
}
```

```python
# Initialize a list to store results
scores = []

# Loop through each model and its parameters
for model_name, mp in model_params.items():
    # Set up GridSearchCV
    clf = GridSearchCV(mp['model'],mp['params'], cv = 5,return_train_score=False)
    # Fit the model
    clf.fit(iris.data,iris.target)
    
    # Store results
    scores.append( {
        'model':model_name, 
        'best_score': clf.best_score_,
        'best_params':clf.best_params_
    })

```

```python
dff = pd.DataFrame(scores, columns=['model','best_score','best_params'])
dff
```
