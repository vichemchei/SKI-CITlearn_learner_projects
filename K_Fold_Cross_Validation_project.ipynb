```python
def get_score(model, X_train,X_test,y_train,y_test):
    model.fit(X_train,y_train)
    return model.score(X_test,y_test)
```

```python
model_L = []
model_svm = []
model_rf = []

from sklearn.model_selection import StratifiedKFold

skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

for train_index, test_index in skf.split(iris.data, iris.target):
    X_train, X_test = iris.data[train_index], iris.data[test_index]
    y_train, y_test = iris.target[train_index], iris.target[test_index]

    # Logistic Regression
    model_L.append(get_score(LogisticRegression(n_jobs=20),X_train,X_test,y_train,y_test))
    model_svm.append(get_score(SVC(kernel='rbf',C = 30, gamma= 'auto'),X_train,X_test,y_train,y_test))
    model_rf.append(get_score(RandomForestClassifier(n_estimators= 40),X_train,X_test,y_train,y_test))
```
