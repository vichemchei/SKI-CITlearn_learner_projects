```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
import numpy as np
from sklearn.datasets import load_digits

digits = load_digits()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(digits.data,digits.target, test_size=0.3)
```

```python
lr = LogisticRegression()
lr.fit(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```

```python
svm= SVC(kernel='rbf',C = 30, gamma= 'auto')
svm.fit(X_train,y_train)
```

```python
svm.score(X_test,y_test)
```

```python
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
```

```python
rf.score(X_test,y_test)
```

```python
from sklearn.model_selection import KFold
kf = KFold (n_splits= 3)
kf
```

```python
for train_index,test_index in kf.split([1,2,3,4,5,6,7,8,9]):
    print(train_index,test_index)
```

```python
def get_score(model, X_train,X_test,y_train,y_test):
    model.fit(X_train,y_train)
    return model.score(X_test,y_test)
```

```python
get_score(LogisticRegression(),X_train,X_test,y_train,y_test)
```

```python
from sklearn.model_selection import StratifiedGroupKFold
folds = StratifiedGroupKFold(n_splits=3)
```

```python
score_L = []
scores_svm = []
score_rf = []
for train_index, test_index in kf.split(digits.data):
    X_train,X_test,y_train,y_test = digits.data[train_index], digits.data[test_index],digits.target[train_index],digits.target[test_index]
    score_L.append(get_score(LogisticRegression(n_jobs=20),X_train,X_test,y_train,y_test))
    scores_svm.append(get_score(SVC(kernel='rbf',C = 30, gamma= 'auto'),X_train,X_test,y_train,y_test))
    score_rf.append(get_score(RandomForestClassifier(n_estimators= 40),X_train,X_test,y_train,y_test))
```
