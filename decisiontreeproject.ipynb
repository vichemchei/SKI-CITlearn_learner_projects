```python
df = pd.DataFrame(data_dict)
df
```

```python
df['salary_above_100k'] = df['salary_above_100k'].astype(int)
df
```

```python
inputs = df.drop('salary_above_100k', axis = 'columns')
target = df['salary_above_100k']

inputs
```

```python
from sklearn.preprocessing import LabelEncoder


# Create separate LabelEncoder objects for each column
le_country = LabelEncoder()
le_education_level = LabelEncoder()
le_position = LabelEncoder()
le_company = LabelEncoder()

# Encode each column and store the results in a new DataFrame (inputs_train)
# inputs_train = df.copy()  # Create a copy of the original DataFrame

inputs['country'] = le_country.fit_transform(df['country'])
inputs['education_level'] = le_education_level.fit_transform(df['education_level'])
inputs['position'] = le_position.fit_transform(df['position'])
inputs['company'] = le_company.fit_transform(df['company'])

inputs.head()
```

```python
target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs,target, test_size=0.2, random_state=42)
X_train
```
